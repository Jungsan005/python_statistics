```
# 인구수 시각화
# symbol 크기는 인구수 대비 비율로 설정

pop = pd.read_csv('seoul_pop_2019.csv', header=None, encoding='euc-kr')

pop.columns=['gu', 'pop']

pop.head(5)

```

```
# 컴마제거, 숫자형으로 변환
pop['pop'] = pop['pop'].str.replace(',','')
pop['pop'] = pop['pop'].apply(lambda x: int(x))

# 두 데이터프레임 합침
df = pd.merge(pop, latlon)
df.head(5)
```

```
# 각 구별 인구수에 따라 심볼의 크기를 조정함
layers = []
fig = gmaps.figure(layout=mylayout, center=seoul, zoom_level=12) # 지도 불러옴

for i in range(len(df)):
  size = int((df.iloc[i,1]*0.0001)/2)  # 인구수
  loc =  [(df.iloc[i,2],df.iloc[i,3])]

  layer = gmaps.symbol_layer(loc, fill_color='red', stroke_color='red', scale=size)
  layers.append(layer) # 심블레이어 생성후 리스트에 저장 

for i in range(len(layers)):  # 생성한 레이어들을 지도에 추가
  fig.add_layer(layers[i])

embed_minimal_html('map2e.html', views=[fig])
display(HTML(filename='map2e.html'))
```

```
# 구별 인구수 검색을 수월하게 하기 위해 
# 인구수 데이터를 dict형으로 변환
df = pop.set_index('gu').to_dict()
df['pop']['종로구']
```

```
# 파이썬에서 사용하는 색상 팔레트
# jiffyclub.github.io/palettable
# 그래프에서 색상을 사용하기 전에 용도/대상을 파악

# qualitative : (질적) 범주형데이터 시각화 시 사용가능 색상
# diverging :  수치, 순서 데이터 시각화 시 사용가능 색상
# sequential : 수치/범주 데이터 시각화 시 사용가능 색상
```

```
# 구글에서 matplotlib colormap로 검색
from matplotlib.cm import viridis   # 노랑 -> 보라
from matplotlib.cm import inferno   # 노랑 -> 검정
from matplotlib.colors import to_hex

print(viridis(0))
print(viridis(1))
print(viridis(0.5))
```

```
# 인구수에 따른 대비 표시할 색상 생성
# pmin : 최소 인구수
# prange: 인구수 범위 (최대~최소)

def catch_color(pop, pmin, prange):
  norm = (pop - pmin) / prange       # 0~1사이 값을 갖도록 정규화함
  invers = 1-norm                   # 색상반전을 위해 1에서 정규화 값을 뺌
  mpcolor = viridis(invers)         # 색상표에 값을 넣어서 적절한 색을 알아냄 
  gmapcolor = to_hex(mpcolor)       # 알아낸 색을 16진수로 변환
  return gmapcolor
```

```
# 인구수에 따라 표사할 색상을 생성하고
# color 라는 리스트에 저장
colors = []

pmin = min(pop['pop']) # 최소인구수
pmax = max(pop['pop']) # 최대인구수
prange = pmax - pmin   # 범위

# seoul_geo_simple.json에서 구 이름 뽑아
# 인구수를 추출한 다음 그것에 대한 색상 알아냄
for gu in seoul_geo['features']:
  guname = gu['properties']['name']  # 구이름 추출
  pop = df['pop'][guname]  # 구별 인구수 추출

  color = catch_color(pop, pmin, prange)
  # 인구수별 색상 추출
  colors.append(color)
```

```
# 추출한 색상으로 인구수 시각화
fig = gmaps.figure(layout=mylayout, center=seoul, zoom_level=10) # 지도 불러옴

layer = gmaps.geojson_layer(seoul_geo, fill_color=colors, stroke_color=colors, fill_opacity=0.45 )

# 마커정보와 위치를 이용해서 마커 생성
fig.add_layer(layer) # 생성한 마커를 레이어로 추가

embed_minimal_html('map3b.html', views=[fig])
display(HTML(filename='map3b.html'))
```
